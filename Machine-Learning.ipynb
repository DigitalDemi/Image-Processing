```python
import numpy as np
import pandas as pd
import random

# Define categories and distributions
genders = ['Male', 'Female', 'Other']
job_roles = ['Engineer', 'Teacher', 'Doctor']
ages = np.random.randint(18, 66, size=1000)  # Ages between 18 and 65
incomes = np.random.uniform(20000, 100000, size=1000)  # Income between $20k and $100k

# Generate unbiased data
data_unbiased = pd.DataFrame({
    'Age': ages,
    'Gender': [random.choice(genders) for _ in range(1000)],
    'Income': incomes,
    'Job Role': [random.choice(job_roles) for _ in range(1000)]
})

print(data_unbiased.head())
```

```python
# Introduce a gender-income bias
biased_incomes = []
for i in range(1000):
    if data_unbiased['Gender'][i] == 'Male':
        income = np.random.uniform(50000, 100000)  # Higher income range for males
    else:
        income = np.random.uniform(20000, 50000)   # Lower income range for females and others
    biased_incomes.append(income)

# Modify the original dataset with biased incomes
data_biased = data_unbiased.copy()
data_biased['Income'] = biased_incomes

print(data_biased.head())
```

```python
biased_job_roles = []
for i in range(1000):
    if data_unbiased['Gender'][i] == 'Male':
        role = np.random.choice(['Engineer', 'Doctor'], p=[0.7, 0.3])
    else:
        role = np.random.choice(['Teacher', 'Engineer'], p=[0.7, 0.3])
    biased_job_roles.append(role)

# Update the biased dataset with job roles
data_biased['Job Role'] = biased_job_roles

print(data_biased.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Income', data=data_biased)
plt.title('Income Distribution by Gender (Biased)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate synthetic data
np.random.seed(0)
X = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = 2 * X + 3 + np.random.randn(100) * 2  # Linear relationship with noise

# 2. Helper functions to fit polynomial models and predict
def poly_features(X, degree):
    """Generate polynomial features for X up to a given degree."""
    return np.vstack([X**i for i in range(degree + 1)]).T

def fit_polynomial(X, y, degree):
    """Fit a polynomial regression model of a given degree."""
    X_poly = poly_features(X, degree)
    weights = np.linalg.pinv(X_poly).dot(y)  # Using pseudo-inverse to solve linear equation
    return weights

def predict_polynomial(X, weights):
    """Predict using the polynomial model with given weights."""
    X_poly = poly_features(X, len(weights) - 1)
    return X_poly.dot(weights)

# 3. Plot function to visualize the results
def plot_model(X, y, y_pred, title):
    plt.scatter(X, y, color='blue', label='True data')
    plt.plot(X, y_pred, color='red', label='Model prediction')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# 4. Underfitting Example: Fit a low-degree polynomial (e.g., degree = 1, simple linear model)
degree_underfit = 1
weights_underfit = fit_polynomial(X, y, degree_underfit)
y_pred_underfit = predict_polynomial(X, weights_underfit)

# Plot the underfitting model
plot_model(X, y, y_pred_underfit, f'Underfitting (Degree = {degree_underfit})')

# 5. Overfitting Example: Fit a high-degree polynomial (e.g., degree = 15)
degree_overfit = 15
weights_overfit = fit_polynomial(X, y, degree_overfit)
y_pred_overfit = predict_polynomial(X, weights_overfit)

# Plot the overfitting model
plot_model(X, y, y_pred_overfit, f'Overfitting (Degree = {degree_overfit})')

# 6. Balanced Example: Fit a medium-degree polynomial (e.g., degree = 3)
degree_balanced = 3
weights_balanced = fit_polynomial(X, y, degree_balanced)
y_pred_balanced = predict_polynomial(X, weights_balanced)

# Plot the balanced model
plot_model(X, y, y_pred_balanced, f'Balanced Fit (Degree = {degree_balanced})')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 2 subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(2)

# Plot the sine wave in the first subplot
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Plot the cosine wave in the second subplot
ax2.plot(x, y2, color='orange')
ax2.set_title('Cosine Wave')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```
